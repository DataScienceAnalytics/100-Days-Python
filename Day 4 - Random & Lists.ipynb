## Rock, paper, scissors project

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
```

```python
gestures = [rock, paper, scissors]
```

```python
your_choice = int(input('What do you choose? Type 0 for rock, 1 for paper and 2 for scissors: '))
computer_choice = random.randint(0,2)
print(f'Your choice is {gestures[your_choice]}')
print(f'Computer choice is {gestures[computer_choice]}')
if your_choice == computer_choice:
    print('Draw')
elif your_choice==0 and computer_choice==1:
    print('Lose')
elif your_choice==0 and computer_choice==2:
    print('Win')
elif your_choice==1 and computer_choice==2:
    print('Lose')
elif your_choice==1 and computer_choice==0:
    print('Win')
elif your_choice==2 and computer_choice==0:
    print('Lose')
elif your_choice==2 and computer_choice==1:
    print('Win')
```
