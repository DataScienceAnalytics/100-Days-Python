## Exercise 1 - Cans of Paint Calculation

```python
test_h = int(input("Height of wall: "))
test_w = int(input("Width of wall: "))
coverage = 5

def paint_calc(height,width,cover):
    print('You need '+ str(int(height*width/cover) + ((height*width) % cover > 0)) + ' cans')

paint_calc(height=test_h, width=test_w, cover=coverage)
```

## Exercise 2 - Prime Number Checker
- Prime number: can only be divided by 1 and itself

```python
n = int(input("Check this number: "))

def prime_checker(number):
    flag = 0
    
    if number % 1==0:
        for i in range(2,number):
            if number%i==0:
                flag = 1
                break
            else:
                flag = 0
        
        if flag == 1 :
            print('This is not a prime number')
        else:
            print('This is a prime number')
        
    else:
        print('This is not a prime number')
```

```python
prime_checker(number=18)
```

## Exercise 3 - Caesar Cipher

```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
text = input("Type your message:\n").lower()
shift = int(input("Type the shift number:\n"))
```

```python
def encode(word, shift):
    encoded_word = ''
    for i in range(len(word)):
        encoded_word += alphabet[alphabet.index(word[i])+shift]
    print(f"The encoded text is {encoded_word}")
```

```python
encode(text, shift)
```

```python
def decode(word, shift):
    encoded_word = ''
    for i in range(len(word)):
        encoded_word += alphabet[alphabet.index(word[i])-shift]
    print(f"The decoded text is {encoded_word}")
```

```python
decode('mjqqt', 5)
```

```python
#Combine both function
def caesar(word, shift, encode_or_decode):
    if encode_or_decode == 'decode':
        shift *= -1
    encoded_word = ''
    
    for i in range(len(word)):
        if word[i].isalpha():
            encoded_word += alphabet[alphabet.index(word[i])+shift]
        else:
            encoded_word += word[i]
    
    print(f"The encoded text is {encoded_word}") 
```

```python
caesar('mjqqt', 5, 'decode')
```

```python
caesar('hello', 5, 'encode')
```
