## Age Left Calculator
Create a program using maths and f-Strings that tells us how many days, weeks, months we have left if we live until 90 years old.

It will take your current age as the input and output a message with our time left in this format:

    You have x days, y weeks, and z months left.

Where x, y and z are replaced with the actual calculated numbers.

Hint
- There are 365 days in a year, 52 weeks in a year and 12 months in a year.
- Try copying the example output into your code and replacing the relevant parts so that the sentence is formated the same way.


```python
age = input("What's your current age? ")
days_left = (90-int(age))*365
weeks_left = (90-int(age))*52
months_left = (90-int(age))*12

print(f"You have {days_left} days, {weeks_left} weeks, and {months_left} months left.")
```

## Tip Calculator


```python
#If the bill was $150.00, split between 5 people, with 12% tip. 

#Each person should pay (150.00 / 5) * 1.12 = 33.6
#Format the result to 2 decimal places = 33.60

#Tip: There are 2 ways to round a number. You might have to do some Googling to solve this.ðŸ’ª

#Write your code below this line ðŸ‘‡
print("Welcome to the tip calculator")
total_bill = float(input("What was the total bill? $"))
tip_pct = int(input("How much tip would you like to give? 10, 12, or 15? "))
number_of_people = int(input('How many people to split the bill? '))

total_per_person = round(total_bill/number_of_people * (1+tip_pct/100),2)
print("Each person should pay: ${:.2f}".format(total_per_person)) #float to string with 2 decimal format
#If we want to display the second decimal places(for example 33.6 to 33.60)
#, we need to format it
```
