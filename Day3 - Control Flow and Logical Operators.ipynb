### Difference between if + if and if + elif

- Multiple if's means your code would go and check all the if conditions, 
- where as in case of elif, if one if condition satisfies it would not check other

```python
#You might get issues if you use if+if+else
#That is because BOTH if statements are being evaluated. 
#The computer sees them as two separate statements: 
''' The first one is true and so "pay 5" is printed.
    The second if statement is false and 
    therefore the else part is executed printing "pay 12"'''

age = int(input('input your age: '))
if age < 12:
    print('pay 5')
if age <= 18:
    print('pay 7')
else:
    print('pay 12')
```

You can see that elif is slightly faster. <br> This would be more apparent if there were more ifs and more elifs.

```python
#Run time difference between if if and if elif
def multipleif(text):
    if text == 'sometext':
        print(text)
    if text == 'nottext':
        print("notanytext")

def eliftest(text):
    if text == 'sometext':
        print(text)
    elif text == 'nottext':
        print("notanytext")

text = "sometext"

timeit multipleif(text)
100000 loops, best of 3: 5.22 us per loop

timeit eliftest(text)
100000 loops, best of 3: 5.13 us per loop
```

### Rule of thumb:
- When using multiple condition. Use <b>elif</b> after if instead of <b>if</b>

## BMI Calculator 2.0
Write a program that interprets the Body Mass Index (BMI) based on a user's weight and height.

It should tell them the interpretation of their BMI based on the BMI value.

- Under 18.5 they are underweight
- Over 18.5 but below 25 they have a normal weight
- Over 25 but below 30 they are slightly overweight
- Over 30 but below 35 they are obese
- Above 35 they are clinically obese.

```python
# ðŸš¨ Don't change the code below ðŸ‘‡
height = float(input("enter your height in m: "))
weight = float(input("enter your weight in kg: "))
# ðŸš¨ Don't change the code above ðŸ‘†

BMI = height/(weight**2)
if BMI < 18.5:
    print('Underweight')
elif BMI < 25:
    print('Normal weight')
elif BMI < 30:
    print('Slightly overweight')
elif BMI < 35:
    print('Obese')
else:
    print('Clinically obese')
```

### Love calculator exercise

```python
# ðŸš¨ Don't change the code below ðŸ‘‡
print("Welcome to the Love Calculator!")
name1 = input("What is your name? \n")
name2 = input("What is their name? \n")
# ðŸš¨ Don't change the code above ðŸ‘†
#Write your code below this line ðŸ‘‡
```

```python
name1 = name1.lower()
name2 = name2.lower()
name_merge = name1+name2

true_count_list = [name_merge.count('true'[i]) for i in range(len('true'))]
true_count = sum(true_count_list)
love_count_list = [name_merge.count('love'[i]) for i in range(len('love'))]
love_count = sum(love_count_list)
```

```python
true_love_index = int(str(true_count)+str(love_count))
if true_love_index < 10 or true_love_index > 90:
    print(f"Your score is {true_love_index}, you go together like coke and mentos.")
elif true_love_index > 40 and true_love_index < 50:
    print(f"Your score is {true_love_index}, you are alright together.")
else:
    print(f"Your score is {true_love_index}.")
```
